```python
# Define the file path of the CSV dataset on Google Drive
file_path = '/content/drive/MyDrive/RAfi/gamers_dataC.csv'
```

```python
# Load the CSV dataset into a pandas DataFrame
df = pd.read_csv(file_path)
```

```python
# Define the attribute names
attributes = ['Age 16-20', 'Age 21-25', 'Age 26-30', 'Age 31-older', 'E-sports', 'casual', 'hardcode',
              'Genre FPS-TPS', 'Genre Role-playing', 'Genre MOBA', 'Genre Sports&simulation',
              'PC', 'Mobile', 'X-BOX', 'offline', 'online',
              'Adrinaline Rush, Anger Issues, Instant Reaction, Getting Tilted',
              'Quick Thinking, Taking Break & Calm, Taking Instant Decision, Thinking Capabilities Improves', 'Short time Memory loss, Headace & Dehazed, Illution happen, Fogetting The runing Tasks',
              'Relaxation,  Mood Enhencement, Focous & concentration, Active and accurate']
```

```python
print(df.columns)
```

```python
# Calculate the count, mean, median, and percentage for each attribute
results = []
for attribute in attributes:
    data = df[attribute]
    yes_count = data.eq('YES').sum()
    no_count = data.eq('NO').sum()
    mean = data.eq('YES').mean()
    median = data.eq('YES').median()
    total_count = yes_count + no_count
    yes_percentage = (yes_count / total_count) * 100
    no_percentage = (no_count / total_count) * 100
    results.append((attribute, yes_count, no_count, mean, median, yes_percentage, no_percentage))
```

```python
# Create the bar chart for YES and NO percentages
yes_percentages = [result[5] for result in results]
no_percentages = [result[6] for result in results]

plt.figure(figsize=(12, 8))
plt.bar(attributes, yes_percentages, label='YES')
plt.bar(attributes, no_percentages, bottom=yes_percentages, label='NO')
plt.xlabel('Attributes')
plt.ylabel('Percentage')
plt.title('Percentage of YES and NO for Each Attribute')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Create a DataFrame with the results
results_df = pd.DataFrame(results,
                          columns=['Attribute', 'Yes Count', 'No Count', 'Yes Percentage', 'No Percentage', 'Mean', 'Median'])
```

```python
# Visualize the percentages
plt.figure(figsize=(12, 8))
plt.bar(results_df['Attribute'], results_df['Yes Percentage'], color='green', label='YES')
plt.bar(results_df['Attribute'], results_df['No Percentage'], color='red', bottom=results_df['Yes Percentage'], label='NO')
plt.xlabel('Attribute')
plt.ylabel('Percentage')
plt.title('Percentage of YES and NO for each Attribute')
plt.legend()
plt.xticks(rotation=45)
```

```python
# Display the mean, median, and count values on the plot
for i, row in results_df.iterrows():
    plt.text(i, row['Yes Percentage'] / 2, f'Mean: {row["Mean"]:.2f}\nMedian: {row["Median"]:.2f}\nCount: {row["Yes Count"]}', ha='center', va='center')

plt.show()
```
